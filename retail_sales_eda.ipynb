## Latihan Tugas 1


```python
# import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# membaca data
data_diskon = pd.read_csv('/content/sample_data/Data_diskon_produk.csv')

print(f'jumlah kolom: {len(data_diskon.columns)}')
print(f'jumlah baris: {len(data_diskon)}')
data_diskon.head()
```

```python
# hapus kolom tidak penting di data
tugas1_clean = data_diskon.drop(columns = ["No", "periksa","Unnamed: 3", "1", "mean", "median", "modus"])

# cek ulang jumlah kolom dan baris
print(f'jumlah kolom: {len(tugas1_clean.columns)}')
print(f'jumlah baris: {len(tugas1_clean)}\n')

# melihat data kembali
tugas1_clean.head()
tugas1_clean.info()
tugas1_clean.describe()

```

```python
# visualisasi histrogram
plt.figure(figsize = (10,6))
plt.hist(tugas1_clean['Diskon Produk (%)'], bins = 30)
plt.xlabel('Diskon Produk')
plt.ylabel('Frekuensi')
plt.title('Histgoram Data Diskon Produk')
plt.show()
```

```python
# visualisasi boxplot
plt.figure(figsize = (10,6))
sns.boxplot(x = 'Diskon Produk (%)', data = tugas1_clean)
plt.title('Boxplot Data Diskon Produk')
plt.show()
```

## Latihan Tugas 2

```python
# load data dan menampilkan hal penting terkait data
tugas2 = pd.read_csv('/content/sample_data/data_tugas_2.csv')

# menghapus kolom "no"
tugas2_clean = tugas2.drop(columns = ["No"])
print(f"jumlah kolom: {len(tugas2_clean.columns)}")
print(f"jumlah baris: {len(tugas2_clean)}\n")
tugas2_clean.head()
tugas2_clean.info()
tugas2_clean.describe()
```

```python
# visualisasi menggunakan scatter plot dengan garis
plt.figure(figsize=(8,4))
sns.regplot(x='Diskon Produk (%)',
                y='Volume Penjualan Harian (unit)',
                data = tugas2_clean,
                scatter_kws = {'alpha':0.5},
                line_kws = {'color':'black'})
plt.title('Distribusi Diskon Produk dan Volume Penjualan Harian (unit)')
plt.show()

# visualisasi menggunakan scatter plot
plt.figure(figsize=(8,4))
sns.scatterplot(x = 'Diskon Produk (%)',
                 y = 'Volume Penjualan Harian (unit)',
                 data = tugas2_clean)
plt.show()
```

## Latihan Tugas 3


```python
# load data dan melihat data
tugas3 = pd.read_csv('/content/sample_data/data_tugas_3.csv')

# menghapus kolom "No"
tugas3_clean = tugas3.drop(columns = "No")

# melihat jumlah baris dan kolom
print(f'Jumlah baris: {len(tugas3_clean)}')
print(f'Jumlah Kolom: {len(tugas3_clean.columns)}\n')

tugas3_clean.head()
tugas3_clean.info()
tugas3_clean.describe()


```

List variabel Daftar variabel:

1.	Jenis Toko → kategorik (nominal)

2.	Diskon Produk (%) → numerik (rasio)

3.	Harga Awal Produk (Rp) → numerik (rasio)

4.	Jumlah Hari Promosi (hari/bulan) → numerik

5.	Volume Penjualan Harian (unit) → numerik (variabel utama / dependen)

## Visualisasi Scatter Plot antar dua Variabel Numerik

##### Diskon Produk (%) <-> Volume Penjualan Harian (unit)

```python
# visualisasi scatter plot variabel Diskon Produk dan Volume Penjualan Harian
sns.scatterplot(x = 'Diskon Produk (%)', y = 'Volume Penjualan Harian (unit)', data = tugas3_clean)
plt.title('Distribusi Diskon Produk dan Volume Penjualan Harian')
plt.show()

# scatter plot dengan garis
sns.regplot(x = 'Diskon Produk (%)',
            y = 'Volume Penjualan Harian (unit)',
            data = tugas3_clean,
            scatter_kws ={'alpha':0.5},
            line_kws ={'color':'black'})
plt.show()



```

##### Jumlah Hari Promosi (hari/bulan) <-> Volume Penjualan Harian (unit)

```python
# scatter plot variabel Jumlah Hari Promosi (hari/bulan) dan Volume Penjualan Harian (unit)
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'Jumlah Hari Promosi (hari/bulan)',
                y = 'Volume Penjualan Harian (unit)',
                data = tugas3_clean)
plt.title('Distribusi Jumlah Hari Promosi dan Volume Penjualan Harian')
plt.show()

plt.figure(figsize=(8,5))
sns.regplot(x = 'Jumlah Hari Promosi (hari/bulan)',
            y = 'Volume Penjualan Harian (unit)',
            data = tugas3_clean,
            scatter_kws = {'alpha':0.5},
            line_kws = {'color':'black'})
plt.show()
```

##### Harga Awal Produk (Rp) <-> Volume Penjualan Harian (unit)

```python
# scatter plot variabel Harga Awal Produk (Rp) dan Volume Penjualan Harian (unit)
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'Harga Awal Produk (Rp)',
                y = 'Volume Penjualan Harian (unit)',
                data = tugas3_clean)
plt.title('Distribusi Harga Awal Produk (Rp) dan Volume Penjualan Harian (unit)')
plt.show()

# visualisasi dengan garis
plt.figure(figsize=(8,5))
sns.regplot(x = 'Harga Awal Produk (Rp)',
            y = 'Volume Penjualan Harian (unit)',
            data = tugas3_clean,
            scatter_kws={'alpha':0.5},
            line_kws = {'color':'black'})
plt.show()
```

##### Diskon Produk (%) <-> Harga Awal Produk (Rp)

```python
# scatter plot variabel Diskon Produk (%) dan Harga Awal Produk (Rp)
plt.figure(figsize=(8, 5))
sns.scatterplot(x = 'Diskon Produk (%)',
                y = 'Harga Awal Produk (Rp)',
                data = tugas3_clean)
plt.title('Distribusi Diskon Produk (%) dan Harga Awal Produk')
plt.show()

plt.figure(figsize=(8, 5))
sns.regplot(x = 'Diskon Produk (%)',
            y = 'Harga Awal Produk (Rp)',
            data = tugas3_clean,
            scatter_kws = {'alpha':0.5},
            line_kws = {'color':'black'})
plt.show()
```

##### Heatmap Korelasi seluruh variabel numerik

```python
# menghapus data yang bukan data numerik
tugas3_heatmap = tugas3.drop(columns = ["No", "Jenis Toko"])

# visualisasikan dengan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(tugas3_heatmap.corr(), annot = True)
plt.title('Korelasi antar Variabel Numerik')
plt.show()
```
